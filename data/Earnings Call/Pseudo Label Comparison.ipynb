```python
import pandas as pd
```

```python
df1 = pd.read_csv('twitter.csv')
df1.rename(columns = {"Unnamed: 0": "Index"}, inplace = True)
df1
```

```python
df2 = pd.read_csv('distilbert.csv')
df2.rename(columns = {"Unnamed: 0": "Index"}, inplace = True)
df2
```

```python
comparison_data = []
for index, row in df1.iterrows():
    
    #刪掉空的column
    summary = row['summary']
    label_file1 = row['label']
    score_file1 = row['score']
    
    #自己新增Index
    df2_row = df2[df2['Index'] == index]
    
    if not df2_row.empty:
        label_file2 = df2_row['label'].iloc[0]
        score_file2 = df2_row['score'].iloc[0]
        
        if label_file1 == label_file2:
            scorefinal = 0.5 * (score_file1 + score_file2)
  
            comparison_data.append({
                                    'summary': summary,
                                    'label': label_file1,
                                    'final_score': scorefinal})
    else:
        continue
comparison_df = pd.DataFrame(comparison_data)

print(comparison_df)
```

```python
comparison_df.to_csv('earningscall_final.csv', index = False)
```
