```python
import pandas as pd
from transformers import pipeline
from tqdm import tqdm 
```

```python
dft = pd.read_csv('20182020.csv')#twitter的
dft
```

```python
new_dft = dft.drop_duplicates()
new_dft['label'] = ''
new_dft['score'] = ''
new_dft
```

```python
new_dft.to_csv('twitter.csv', index=False)
```

```python
new_dft = pd.read_csv('twitter.csv')
new_dft
```

```python
mt = pipeline("text-classification", model="cardiffnlp/twitter-roberta-base-sentiment-latest")
```

```python
i=0
for line in tqdm(new_dft['summary'], total=len(new_dft)):
    if isinstance(line, str):
        res = mt(line)
        new_dft.loc[i, 'label'] = res[0]['label']
        new_dft.loc[i, 'score'] = res[0]['score']
    else:
        print(f"Ignoring invalid data: {line}")
    i += 1
    
print("done")
```

```python
new_dfd = pd.read_csv('distilbert.csv')
new_dfd
```

```python
md = pipeline("text-classification", model="austinmw/distilbert-base-uncased-finetuned-tweets-sentiment")
```

```python
i=0
for line in tqdm(new_dfd['summary'], total=len(new_dfd)):
    if isinstance(line, str):
        res = md(line)
        new_dfd.loc[i, 'label'] = res[0]['label']
        new_dfd.loc[i, 'score'] = res[0]['score']
    else:
        print(f"Ignoring invalid data: {line}")
    i += 1
    
print("done")
```
